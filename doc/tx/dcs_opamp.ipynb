```python
from pygmid import Lookup as lk
import numpy as np

# Load 1.2V core lookup tables
lv_nmos = lk('../lib/sg13_lv_nmos.mat')
lv_pmos = lk('../lib/sg13_lv_pmos.mat')

# Design Specs
I_TAIL = 3.5e-3
I_BRANCH = I_TAIL / 2  # Current per branch at equilibrium

print("--- 1.2V LVDS DCS Sizing ---")

# ==========================================
# 1. Tail NMOS (M1)
# ==========================================
gmid_m1 = 12
L_m1 = 0.5      # Typical longer L for current sources
vds_m1 = 0.6    # Characterization at VDD/2

id_w_m1 = lv_nmos.lookup('ID_W', GM_ID=gmid_m1, L=L_m1, VDS=vds_m1)
vgs_m1  = lv_nmos.lookup('VGS', GM_ID=gmid_m1, L=L_m1, VDS=vds_m1)

# Pygmid returns arrays; we use [0] to extract the scalar
w_m1 = I_TAIL / id_w_m1

print(f"M1 (Tail NMOS):")
print(f"W   = {w_m1:.2f} um")
print(f"L   = {L_m1} um")
print(f"VGS = {vgs_m1:.3f} V\n")

# ==========================================
# 2. Differential Pair (M2, M3)
# ==========================================
gmid_sw = 8
L_sw = 0.13     # Minimum L for maximum speed (ft)
vds_sw = 0.6    

id_w_sw = lv_nmos.lookup('ID_W', GM_ID=gmid_sw, L=L_sw, VDS=vds_sw)
vgs_sw  = lv_nmos.lookup('VGS', GM_ID=gmid_sw, L=L_sw, VDS=vds_sw)

w_sw = I_BRANCH / id_w_sw[0]

print(f"M2/M3 (Diff Pair):")
print(f"W   = {w_sw:.2f} um")
print(f"L   = {L_sw} um")
print(f"VGS = {vgs_sw[0]:.3f} V\n")

# ==========================================
# 3. Top PMOS Sources (M4, M5)
# ==========================================
gmid_p = 12
L_p = 0.5
vds_p = 0.6

id_w_p = lv_pmos.lookup('ID_W', GM_ID=gmid_p, L=L_p, VDS=vds_p)
vsg_p  = lv_pmos.lookup('VGS', GM_ID=gmid_p, L=L_p, VDS=vds_p)

w_p = I_BRANCH / id_w_p[0]

print(f"M4/M5 (Top PMOS):")
print(f"W   = {w_p:.2f} um")
print(f"L   = {L_p} um")
print(f"VSG = {vsg_p[0]:.3f} V")
```
